```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
stu = pd.read_csv('ScholarshipApplicationDataFilePretend.csv' , sep=',', encoding='latin1')
```

```python
stu.head()
```

```python
records = []  
for i in range(0, 380):  
    records.append([str(stu.values[i,j]) for j in range(0, 11)])
```

```python
#14 results
#results = list(apriori(stu, min_support=0.0003, min_confidence=0.1, min_lift=2, min_length=2))
#12 results
#results = list(apriori(stu, min_support=0.0005, min_confidence=0.2, min_lift=2, min_length=2))
#11 rules    
results = list(apriori(records, min_support=0.05, min_confidence=0.1, min_lift=2, min_length=2))
   
for item in results:
    hypo = ''.join([x+' ' for x in item.ordered_statistics[0].items_base])
    conc = ''.join([x+' ' for x in item.ordered_statistics[0].items_add])
    print(str(hypo)+ " --> "+str(conc))
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("-----------------------------------")
    
for i in range(0, len(results)):
    result = results[i]
    hypo = ''.join([x+' ' for x in result.ordered_statistics[0].items_base])
    conc = ''.join([x+' ' for x in result.ordered_statistics[0].items_add])
    print(str(hypo)+ " --> "+str(conc))
```
